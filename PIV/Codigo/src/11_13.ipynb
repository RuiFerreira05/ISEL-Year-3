```python
import numpy as np
import cv2
```

```python
cam = cv2.VideoCapture(0)

_, init_frame = cam.read()

old_frame = cv2.cvtColor(init_frame, cv2.COLOR_BGR2GRAY)

median_values = []

while True:
    _, frame = cam.read()
    
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    median_values.append(np.mean(np.abs(frame - old_frame)))
    
    if len(median_values) > 1000:
        median_values.pop(0)

    cv2.putText(frame, f'median: {np.mean(median_values)}', [0, 30], cv2.FONT_HERSHEY_COMPLEX, 0.5, [0, 0, 0])
    cv2.imshow('Camera', frame)

    old_frame = frame
    
    # Press 'q' to exit the loop
    if cv2.waitKey(1) == ord('q'):
        cam.release()
        cv2.destroyAllWindows()
        break
```
