```python
X = rd.randn(2, 1000)
X.shape
```

```python
plt.plot(X[0,:], X[1,:], '.')
plt.axis('scaled')
plt.grid(True)
```

```python
m = np.mean(X, axis=1)
print(np.sum(X, axis=1) / X.shape[1])
print(m)
```

```python
np.cov(X)
```

```python
Xn = (X.T-m).T
print(Xn.shape)
# or
Xn = X - m[:,np.newaxis]
print(Xn.shape)
```

```python
np.mean(Xn, axis=1) 
```

```python
A = 1/np.sqrt(2)*np.array([[1/3,1], [-1/3,1]])
b=np.array([3, 2])

Y = np.dot(A,X) + b[:,np.newaxis]

plt.plot(X[0,:], X[1,:], '.', Y[0,:], Y[1,:], '.')
plt.axis('scaled')
plt.grid(True)
```
