## Matrizes de Distâncias

#### Matriz de distância de dados sintéticos: $\texttt{LAB2distancias001.p}$

```python
fName='data/LAB2distancias001.p'
D=pickle.load(open(fName,'rb'))
```

```python
X=D['X']
y=D['idxClass']
print(X.shape)
print(y.shape)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.axis('scaled')
plt.grid(True)
```

```python
D2=spd.squareform(spd.pdist(X.T,'euclidean'))
D1=spd.squareform(spd.pdist(X.T,'cityblock'))
Dc=spd.squareform(spd.pdist(X.T,'cosine'))
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.imshow(D2)
plt.axis('off')
plt.subplot(132)
plt.imshow(D1)
plt.axis('off')
plt.subplot(133)
plt.imshow(Dc)
plt.axis('off')
```

```python
nx=np.sqrt(np.sum(X**2,axis=0))
Xn=X/nx
Dc2=1-np.dot(Xn.T,Xn)
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(Dc)
plt.axis('off')
plt.subplot(122)
plt.imshow(Dc2)
plt.axis('off')
```

## Classificadores de distância ao centroide

#### Classificador de distância ao centroide com dados sintéticos: $\texttt{Lab2distancias002.p}$

```python
fName='LAB2distancias002.p'
D=pickle.load(open(fName,'rb'))
```

```python
X=D['X']
y=D['idxClass']
print(X.shape)
print(y.shape)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.axis('scaled')
```

```python
#m1=np.mean(X[:,y==1],axis=1)
#m2=np.mean(X[:,y==2],axis=1)
#m3=np.mean(X[:,y==3],axis=1)
m1=np.array([-2,2])
m2=np.array([2,-2])
m3=np.array([0,0])
```

```python
D1=np.sqrt(np.sum((X-m1[:,np.newaxis])**2,axis=0))
D2=np.sqrt(np.sum((X-m2[:,np.newaxis])**2,axis=0))
D3=np.sqrt(np.sum((X-m3[:,np.newaxis])**2,axis=0))
```

```python
D=np.vstack((D1,D2,D3))
ye=np.argmin(D,axis=0)+1
```

```python
from sklearn.metrics import confusion_matrix
MC=confusion_matrix(y,ye)
print(MC)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.plot(X[0,y!=ye],X[1,y!=ye],'ok',alpha=.2)
plt.axis('scaled')
```

```python
m1=np.array([-2,2])
m2=np.array([2,-2])
m3=np.array([0,0])
S1=np.eye(2)*1/4
S2=np.array([[1/4,0],[0,1/4]])
S3=np.array([[3,0],[0,1/10]])

```

```python
from numpy.linalg import pinv
S1i=pinv(S1)
S2i=pinv(S2)
S3i=pinv(S3)
```

```python
D1=np.sqrt(np.sum((X-m1[:,np.newaxis])*np.dot(S1i,X-m1[:,np.newaxis]),axis=0))
D2=np.sqrt(np.sum((X-m2[:,np.newaxis])*np.dot(S2i,X-m2[:,np.newaxis]),axis=0))
D3=np.sqrt(np.sum((X-m3[:,np.newaxis])*np.dot(S3i,X-m3[:,np.newaxis]),axis=0))
D=np.vstack((D1,D2,D3))
ye=np.argmin(D,axis=0)+1
MC=confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.plot(X[0,y!=ye],X[1,y!=ye],'ok',alpha=.2)
plt.axis('scaled')
```

#### Classificador de distância ao centroide com dados reais: Iris Dataset

```python
from sklearn import datasets as DT
D=DT.load_iris()
X=D.data.T
y=D.target
```

```python
m0=np.mean(X[:,y==0],axis=1)
m1=np.mean(X[:,y==1],axis=1)
m2=np.mean(X[:,y==2],axis=1)
D0=np.sqrt(np.sum((X-m0[:,np.newaxis])**2,axis=0))
D1=np.sqrt(np.sum((X-m1[:,np.newaxis])**2,axis=0))
D2=np.sqrt(np.sum((X-m2[:,np.newaxis])**2,axis=0))
D=np.vstack((D0,D1,D2))
ye=np.argmin(D,axis=0)
```

```python
MC=confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```

## Classificador dos k-vizinhos mais próximos

```python
from sklearn.neighbors import KNeighborsClassifier
fName='LAB2distancias002.p'
D=pickle.load(open(fName,'rb'))
X=D['X']
y=D['idxClass']
```

```python
X1a=X[:,:100]
X1b=X[:,500:1000]
X2a=X[:,1000:1100]
X2b=X[:,1500:2000]
X3a=X[:,2000:2100]
X3b=X[:,2500:3000]
```

```python
X1=np.hstack((X1a,X2a,X3a)).T
X2=np.hstack((X1b,X2b,X3b)).T
y1=np.hstack((np.zeros(100),np.ones(100),np.ones(100)*2))
y2=np.hstack((np.zeros(500),np.ones(500),np.ones(500)*2))
```

```python
knn=KNeighborsClassifier(n_neighbors=1).fit(X1,y1)
y2e=knn.predict(X2)
print(np.sum(y2e!=y2))
print(confusion_matrix(y2,y2e))
```

```python
plt.plot(X1b[0,:],X1b[1,:],'.')
plt.plot(X2b[0,:],X2b[1,:],'.')
plt.plot(X3b[0,:],X3b[1,:],'.')
plt.plot(X2[y2!=y2e,0],X2[y2!=y2e,1],'ok',alpha=.5)
```

```python
knn=KNeighborsClassifier(n_neighbors=5).fit(X1,y1)
y2e=knn.predict(X2)
print(np.sum(y2e!=y2))
print(confusion_matrix(y2,y2e))
```

```python
plt.plot(X1b[0,:],X1b[1,:],'.')
plt.plot(X2b[0,:],X2b[1,:],'.')
plt.plot(X3b[0,:],X3b[1,:],'.')
plt.plot(X2[y2!=y2e,0],X2[y2!=y2e,1],'ok',alpha=.5)
```
